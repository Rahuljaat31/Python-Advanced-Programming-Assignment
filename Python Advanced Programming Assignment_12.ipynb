Q1. For this challenge, forget how to add two numbers together. The best explanation on what to do for this function is this meme:


Examples

meme_sum(26, 39)  515

    # 2+3 = 5, 6+9 = 15

    # 26 + 39 = 515

meme_sum(122, 81)  1103

    # 1+0 = 1, 2+8 = 10, 2+1 = 3

    # 122 + 81 = 1103

meme_sum(1222, 30277)  31499

```python
#Ans.1
def meme_sum(x,y):
    x,y = str(x),str(y)
    res = ''
    while len(x) != len(y):
        if len(x) < len(y):
            x ='0'+x
        else:
            y='0'+y
    for i in range(len(x)):
        res += str(int(x[i])+int(y[i]))
    print(f'meme_sum{x,y}  {res}')
```

Q2. Given an integer, create a function that returns the next prime. If the number is prime, return the number itself.
Examples

next_prime(12)  13

next_prime(24)  29

next_prime(11)  11

    # 11 is a prime, so we return the number itself.

```python
#Ans.2 
def next_prime(n):
    nc = n
    while True:
        if (n-1)%6 == 0 or (n+1)%6 ==0 :
            t = n
            for i in range(3,n+1):
                if n%i == 0 and i != n:
                    n = n+1
                    break
            if t == n:
                break
        else:
            n += 1
    
    print(f'next_prime({nc})  {n}')
```

```python
next_prime(12)
```

```python
next_prime(24)
```

```python
next_prime(11)
```

Q3. If a person traveled up a hill for 18mins at 20mph and then traveled back down the same path at 60mph then their average speed traveled was 30mph. Write a function that returns the average speed traveled given an uphill time, uphill rate and a downhill rate. Uphill time is given in minutes. Return the rate as an integer (mph). No rounding is necessary.

Examples

ave_spd(18, 20, 60)  30

ave_spd(30, 10, 30)  15

ave_spd(30, 8, 24)  12

```python
#Ans.3 
def ave_spd(t,s,r):
    dis = s*(t/60)
    dt = dis/r
    res = (2*dis)/((t/60)+dt)
    print(f'ave_spd{t,s,r}  {int(res)}')
```

```python
ave_spd(18, 20, 60)
```

```python
ave_spd(30, 10, 30)
```

```python
ave_spd(30, 8, 24) 
```

Q4. The Kempner Function, applied to a composite number, permits to find the smallest integer greater than zero whose factorial is exactly divided by the number.

kempner(6)  3

1! = 1 % 6 > 0

2! = 2 % 6 > 0

3! = 6 % 6 === 0

kempner(10)  5

1! = 1 % 10 > 0

2! = 2 % 10 > 0

3! = 6 % 10 > 0

4! = 24 % 10 > 0

5! = 120 % 10 === 0

A Kempner Function applied to a prime will always return the prime itself.

kempner(2)  2

kempner(5)  5

Given an integer n, implement a Kempner Function.

Examples:

kempner(6)  3

kempner(10)  5

kempner(2)  2

```python
#Ans.4
def kempner(n):
    def fact(n):
        if n == 1:
            return 1
        else:
            return n * fact(n-1)
    for i in range(1,n+1):
        if fact(i)%n == 0:
            res = i
            break
    print(f'kempner({n})  {res}')
```

```python
kempner(6)
```

```python
kempner(2)
```

```python
kempner(5)
```

```python
kempner(10)
```

Q5. You work in a factory, and your job is to take items from a conveyor belt and pack them into boxes. Each box can hold a maximum of 10 kgs. Given a list containing the weight (in kg) of each item, how many boxes would you need to pack all of the items?

Example

boxes([2, 1, 2, 5, 4, 3, 6, 1, 1, 9, 3, 2])  5

    # Box 1 = [2, 1, 2, 5] (10kg)
    # Box 2 = [4, 3] (7kg)
    # Box 3 = [6, 1, 1] (8kg)
    # Box 4 = [9] (9kg)
    # Box 5 = [3, 2] (5kg)

```python
#Ans.5
def boxes(l):
    lc = l.copy()
    res = []
    t = []
    while True:
        if len(l) != 0:
            if sum(t) <= 10:
                t.append(l.pop(0))
            else:
                l.insert(0,t.pop())
                res.append(t)
                t = []
        else:
            res.append(t)
            t = []
            break
    print(f'boxes({lc})  {len(res)}')
# We need 5 boxes    
```
