Q1. Create a function that takes a list and returns a new list containing only prime numbers.

Examples

filter_primes([7, 9, 3, 9, 10, 11, 27]) ➞ [7, 3, 11]

filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70]) ➞ [10007, 1009]

filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093,
1097]) ➞ [1009, 3, 61, 1087, 1091, 1093, 1097]

```python
#Ans.1 
def filter_primes(l):
    res = []
    for i in l:
        if i in [2,3]:
            res.append(i)
        elif ((i+1)%6 == 0) or ((i-1)%6 == 0) and i != 1:
            res.append(i)
    t = res.copy()
    for i in t:
        for n in range(2,i):
            if i%n == 0:
                res.remove(i)
                break        
    print(f'filter_primes({l}) ➞ {res}')
```

```python
filter_primes([7, 9, 3, 9, 10, 11, 27])
```

```python
filter_primes([10007, 1009, 1007, 27, 147, 77, 1001, 70])
```

```python
filter_primes([1009, 10, 10, 10, 3, 33, 9, 4, 1, 61, 63, 69, 1087, 1091, 1093, 1097])
```

Q2. Once a water balloon pops, is soaks the area around it. The ground gets drier the further away you travel from the balloon.
The effect of a water balloon popping can be modeled using a list. Create a function that takes a list which takes the pre-pop state and returns the state after the balloon is popped. The pre-pop state will contain at most a single balloon, whose size is represented by the only non-zero element.

Examples

pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) ➞ [0, 1, 2, 3, 4, 3, 2, 1, 0]

pop([0, 0, 0, 3, 0, 0, 0]) ➞ [0, 1, 2, 3, 2, 1, 0]

pop([0, 0, 2, 0, 0]) ➞ [0, 1, 2, 1, 0]

pop([0]) ➞ [0]

```python
#Ans.2 
def pop(l):
    num = sorted(l,reverse=True)[0]
    res = []
    if num == 0:
        res.append(0)
    else:
        res.extend([x for x in range((len(l)-1)//2)])
        res.append(num)
        res.extend([((len(l)-2)//2)-x for x in range((len(l)-1)//2)])
    print(f'pop({l}) ➞ {res}')
```

```python
pop([0, 0, 0, 0, 4, 0, 0, 0, 0]) 
```

```python
pop([0, 0, 0, 3, 0, 0, 0]) 
```

```python
pop([0, 0, 2, 0, 0])
```

```python
pop([0]) 
```

Q3. "Loves me, loves me not" is a traditional game in which a person plucks off all the petals of a flower one by one, saying the phrase "Loves me" and "Loves me not" when determining whether the one that they love, loves them back.
Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.

Examples:

loves_me(3) ➞ "Loves me, Loves me not, LOVES ME"

loves_me(6) ➞ "Loves me, Loves me not, Loves me, Loves me not, Loves me, LOVES ME NOT"

loves_me(1) ➞ "LOVES ME"

```python
#Ans.3 
def loves_me(n):
    res = []
    for i in range(n):
        if i%2 ==0:
            res.append('Loves me')
        else:
            res.append('Loves me not')
    res[-1] = res[-1].upper()
    print(f'loves_me({n}) ➞ {", ".join(res)}')
```

```python
loves_me(3)
```

```python
loves_me(6) 
```

```python
loves_me(1) 
```

Q4. Write a function that sorts each string in a list by the letter in alphabetic ascending order (a-z).

Examples:

sort_by_letter(["932c", "832u32", "2344b"]) ➞ ["2344b", "932c", "832u32"]

sort_by_letter(["99a", "78b", "c2345", "11d"]) ➞ ["99a", "78b", "c2345", "11d"]

sort_by_letter(["572z", "5y5", "304q2"]) ➞ ["304q2", "5y5", "572z"]

sort_by_letter([]) ➞ []

```python
#Ans.4 
def sort_by_letter(l):
    t = []
    res = []
    for i in l:
        for j in i:
            if j.isalpha():
                t.append(j)
    t.sort()
    for j in t:
        for i in l:
            if j in i:
                res.append(i)
    
    print(f'sort_by_letter({l}) ➞ {res}')
```

```python
sort_by_letter(["932c", "832u32", "2344b"]) 
```

```python
sort_by_letter(["99a", "78b", "c2345", "11d"])
```

```python
sort_by_letter(["572z", "5y5", "304q2"]) 
```

```python
sort_by_letter([])
```
